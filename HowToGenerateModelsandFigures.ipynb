# How to Generate Models and Figures

There are several steps required to create models and to re-create the figures and they should be run in the order indicated as each step creates files that are required for the next. 


1. Use the Network.py code to train a network with your desired specifications. This will generate a file named by the configuration selected that will contain both the initialized weights as well as the trained weights and final accuracies over training epochs. By default the model will train for 100 epochs on the sMNIST data set, unless --nepochs n option is set.  Networks can have any of a number of configurations ex; 

python Network.py --ConnType WattsStrogatz --ModelType RNN --nhidden 198 --ninputs 28 --nlayers 1 --batch_size 256 --nNN 128 --p 0.5 --init_gain 1.5 --suffix Run_1

or for biologically realistc mesoscopic model (optionally threshold connectivity to match sparsity of V1 model):

python Network.py --ConnType Density --ModelType RNN --nhidden 198 --ninputs 28 --nlayers 1 --batch_size 256 --init_gain 0.5 --thresh_to 5573 --suffix Run_2

or for biologically realistc model V1 with connections from layers 2/3 and 4 and training with optional Dale's Law:

python Network.py --ConnType v1dd --ModelType RNN --nhidden 198 --ninputs 28 --nlayers 1 --batch_size 256 --init_gain 2.5 --suffix 23_4 --Dales --suffix Run_3

where --p is for rewiring probabilty and the suffix allows you to run multiple iniitalilzations of the same model with output to different filenames with the added suffix. See code for more model options.


2. Calculate the Lyapunov Exponents on the models - first on the pre-trained weights, using lyapunov_Vogt.py. Then run on the trained models. This code will create a file for each model (pre/post training) containing the Lyapunov Exponents.


3. Run plot_jacobian_product.py which will generate some plots, but importantly will also generate a csv file with the transition points for each model variant averaged over repeated runs which is required for creation of subsequent figures.


4. Generate figures related to model performance (accuracy and # of epochs to X% accuracy) using MaxAcc_afo_Gain.py and time_to_max_acc_aafo_gain.py


5. Figures for rich vs lazy learning can be generated using the codes rich_lazy_perprob.py (computes the normed weight changes for each model and genrates figures) and KA_RA.py (computes the kernel alignment and representational alignment and generates figures).


6. All other figures in the main manuscript (except some in the supplemental material) can be genrated using EM_nNN28_summary.py which has 3 different opotions: MESOSCOPIC, SUPPLEMENTARY, V1DD which make the plots for the figures related to each Figures 6,7 and S5.


7. Supplemental figures of the eigenspectrum are generated by distroII.py and those for the Lyapunov spectrum were generated in step 3 with the use of plot_jacobian_product. One figure is generated per each model file.
